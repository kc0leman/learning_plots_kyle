```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
spotify_df = pd.read_csv("../data/Cleaned_Spotify_2024_Global_Streaming_Data.csv")
spotify_df.head()
```

```python
### BAR PLOTS ###
```

```python
# group by genre and calculate the average total streams
avg_streams = spotify_df.groupby('Genre')['Total Streams (Millions)'].mean().sort_values(ascending=False)
avg_streams
```

```python
plt.bar(avg_streams.index, avg_streams.values)
plt.show()
```

```python
plt.bar(avg_streams.index, avg_streams.values)

plt.xlabel('Genre')
plt.ylabel('Average Total Streams (Millions)')
plt.title('Average Total Streams by Genre')

plt.show()
```

```python
# can change figure size by using plt.figure(figsize=(x,x))

plt.figure(figsize=(10,6))
plt.bar(avg_streams.index, avg_streams.values)

plt.xlabel('Genre')
plt.ylabel('Average Total Streams (Millions)')
plt.title('Average Total Streams by Genre')
# change the angle of the text by using plt.xticks      ha is horizontal alignment
plt.xticks(rotation=45, ha='right')
# plt.tight_layout makes sure that everything has the correct spacing around it and is easily viewable by all devices that are running it
plt.tight_layout()
plt.show()
```

```python
# a figure can have one or more axes objects    # the borders outside the chart are called spines

plt.figure(figsize=(10,6))
plt.bar(avg_streams.index, avg_streams.values)
# plt.gca() gca is "get current axes"; when you call plt.gca() you are telling matplotlib to give you the axes object for the current figure so you can work with it directly
ax = plt.gca()
ax.spines['top'].set_visible(False)   # essentially says you don't want to see this line
ax.spines['right'].set_visible(False)

plt.xlabel('Genre')
plt.ylabel('Average Total Streams (Millions)')
plt.title('Average Total Streams by Genre')

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
avg_artist_streams = spotify_df.groupby('Artist')['Total Streams (Millions)'].mean().sort_values(ascending=False)
avg_artist_streams
```

```python
plt.figure(figsize=(10,6))

# you can create a list of colors
colors = ['#fa9fb5']*len(avg_artist_streams)     # this will make a list of specified color for the number of bars
# plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width=0.8)

max_index = avg_artist_streams.idxmax()     # will give you the index of the maximum value in the list of values
highlight_position = avg_artist_streams.index.get_loc(max_index)     # this will find the index and put it into colors in this instance
colors[highlight_position] = '#f030ad'

# plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = '#fa9fb5', width=0.8)     # You can find certain colors by using resources like colorbrewer.com, coolors.com, the html color codes, etc
plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width=0.8)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel('Artist')
plt.ylabel('Average Total Streams (Millions)')
plt.title('Average Total Streams by Artist')

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
# Filtering for specific artist

plt.figure(figsize=(10,6))

special_artist = 'Ariana Grande'
colors = ['#f030ad' if artist == special_artist else 'gray' for artist in avg_artist_streams.index]

plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width=0.8)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel('Artist')
plt.ylabel('Average Total Streams (Millions)')
plt.title('Average Total Streams by Artist')

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
# Change title name

plt.figure(figsize=(10,6))

special_artist = 'Ariana Grande'
colors = ['#f030ad' if artist == special_artist else 'gray' for artist in avg_artist_streams.index]

plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width=0.8)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel('Artist')
plt.ylabel('Average Total Streams (Millions)')
plt.title('Ariana Grande Ranks 2nd in Total Average Streams')

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
plt.scatter(spotify_df['Streams Last 30 Days (Millions)'], spotify_df['Total Streams (Millions)'])

plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.scatter(
    spotify_df['Streams Last 30 Days (Millions)'], 
    spotify_df['Total Streams (Millions)'],
    color = '#08cc15',     # can type '#ffffff' (6 f characters) to bring up customizable color palette)
    alpha = 0.8,     # can also change transparency by using alpha
    s=25     # s is the point size (the dot size)
    )
plt.xlabel('Streams this Month')
plt.ylabel('Total Streams')
plt.title('No Apparent Correlation Between Monthly Listens and Total Streams')

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
# can be used if you want to highlight only certain points; eg. only look at values for a certain genre (in this instance it is 'Pop')

plt.figure(figsize=(10,6))

featured_genre = 'Pop'

colors = ['#08cc15' if genre == featured_genre else '#d1d5d1' for genre in spotify_df['Genre']]     # must tell what a genre is as the variable 'genre' is different from the 'Genre' variable in the dataframe

plt.scatter(
    spotify_df['Streams Last 30 Days (Millions)'], 
    spotify_df['Total Streams (Millions)'],
    color = colors,
    alpha = 0.8,
    s=25
    )
plt.xlabel('Streams this Month')
plt.ylabel('Total Streams')
# plt.title('No Apparent Correlation Between Monthly Listens and Total Streams')

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
# can change the size of the font of the title; can also change the title name, labels, among many other things from the axes object
ax.set_title('Do Pop Songs Achieve Higher Monthly Listens or Total Streams?', fontsize = 18, pad=35)

# for loop is faster and less coding, but can also write the values for both linewidth and alpha for 'top' and 'right'
for spine in ax.spines.values():
    spine.set_linewidth(0.25)
    spine.set_alpha(0.5)

plt.tight_layout()
plt.show()
```
