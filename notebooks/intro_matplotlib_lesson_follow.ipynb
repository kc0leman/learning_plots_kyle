```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
spotify_df = pd.read_csv("../data/Cleaned_Spotify_2024_Global_Streaming_Data.csv")
spotify_df.head()
```

```python
### BAR PLOTS ###
```

```python
# group by genre and calculate the average total streams
avg_streams = spotify_df.groupby('Genre')['Total Streams (Millions)'].mean().sort_values(ascending=False)
avg_streams
```

```python
plt.bar(avg_streams.index, avg_streams.values)
plt.show()
```

```python
plt.bar(avg_streams.index, avg_streams.values)

plt.xlabel('Genre')
plt.ylabel('Average Total Streams (Millions)')
plt.title('Average Total Streams by Genre')

plt.show()
```

```python
# can change figure size by using plt.figure(figsize=(x,x))

plt.figure(figsize=(10,6))
plt.bar(avg_streams.index, avg_streams.values)

plt.xlabel('Genre')
plt.ylabel('Average Total Streams (Millions)')
plt.title('Average Total Streams by Genre')
# change the angle of the text by using plt.xticks      ha is horizontal alignment
plt.xticks(rotation=45, ha='right')
# plt.tight_layout makes sure that everything has the correct spacing around it and is easily viewable by all devices that are running it
plt.tight_layout()
plt.show()
```

```python
# a figure can have one or more axes objects    # the borders outside the chart are called spines

plt.figure(figsize=(10,6))
plt.bar(avg_streams.index, avg_streams.values)
# plt.gca() gca is "get current axes"; when you call plt.gca() you are telling matplotlib to give you the axes object for the current figure so you can work with it directly
ax = plt.gca()
ax.spines['top'].set_visible(False)   # essentially says you don't want to see this line
ax.spines['right'].set_visible(False)

plt.xlabel('Genre')
plt.ylabel('Average Total Streams (Millions)')
plt.title('Average Total Streams by Genre')

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
avg_artist_streams = spotify_df.groupby('Artist')['Total Streams (Millions)'].mean().sort_values(ascending=False)
avg_artist_streams
```

```python
plt.figure(figsize=(10,6))

# you can create a list of colors
colors = ['#fa9fb5']*len(avg_artist_streams)     # this will make a list of specified color for the number of bars
# plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width=0.8)

max_index = avg_artist_streams.idxmax()     # will give you the index of the maximum value in the list of values
highlight_position = avg_artist_streams.index.get_loc(max_index)     # this will find the index and put it into colors in this instance
colors[highlight_position] = '#f030ad'

# plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = '#fa9fb5', width=0.8)     # You can find certain colors by using resources like colorbrewer.com, coolors.com, the html color codes, etc
plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width=0.8)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel('Artist')
plt.ylabel('Average Total Streams (Millions)')
plt.title('Average Total Streams by Artist')

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
# Filtering for specific artist

plt.figure(figsize=(10,6))

special_artist = 'Ariana Grande'
colors = ['#f030ad' if artist == special_artist else 'gray' for artist in avg_artist_streams.index]

plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width=0.8)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xlabel('Artist')
plt.ylabel('Average Total Streams (Millions)')
plt.title('Average Total Streams by Artist')

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.scatter(
    spotify_df['Streams Last 30 Days (Millions)'], 
    spotify_df['Total Streams (Millions)'],
    color = '#08cc15',     # can type '#ffffff' (6 f characters) to bring up customizable color palette)
    alpha = 0.8,     # can also change transparency by using alpha
    s=25     # s is the point size (the dot size)
    )
plt.xlabel('Streams this Month')
plt.ylabel('Total Streams')
plt.title('No Apparent Correlation Between Monthly Listens and Total Streams')

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
# can be used if you want to highlight only certain points; eg. only look at values for a certain genre (in this instance it is 'Pop')

plt.figure(figsize=(10,6))

featured_genre = 'Pop'

colors = ['#08cc15' if genre == featured_genre else '#d1d5d1' for genre in spotify_df['Genre']]     # must tell what a genre is as the variable 'genre' is different from the 'Genre' variable in the dataframe

plt.scatter(
    spotify_df['Streams Last 30 Days (Millions)'], 
    spotify_df['Total Streams (Millions)'],
    color = colors,
    alpha = 0.8,
    s=25
    )
plt.xlabel('Streams this Month')
plt.ylabel('Total Streams')
# plt.title('No Apparent Correlation Between Monthly Listens and Total Streams')

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
# can change the size of the font of the title; can also change the title name, labels, among many other things from the axes object
ax.set_title('Do Pop Songs Achieve Higher Monthly Listens or Total Streams?', fontsize = 18, pad=35)

# for loop is faster and less coding, but can also write the values for both linewidth and alpha for 'top' and 'right'
for spine in ax.spines.values():
    spine.set_linewidth(0.25)
    spine.set_alpha(0.5)

plt.tight_layout()
plt.show()
```

```python
### LINE CHARTS ###

# line charts are very good for showing change over time
```

```python
yearly_avg = spotify_df.groupby('Release Year')['Total Streams (Millions)'].mean()
yearly_avg
```

```python
yearly_avg = spotify_df.groupby('Release Year')['Total Streams (Millions)'].mean().round(2)     # can use .round() to show a rounded number by putting the desired place values inside the parenthesis
yearly_avg
```

```python
plt.figure(figsize=(10,6))
plt.plot(yearly_avg.index, yearly_avg.values)     # to plot, you insert the x and y axes here

# steps if you want to change where the axes starts
plt.ylim(0, 3500)

plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.plot(
    yearly_avg.index, 
    yearly_avg.values,
    marker='x',     # you can set a marker to what you would like your plots to be
    linestyle = '--',    # can change the style of the line (solid, dashed, etc.)
    color = '#20a32b',     # can change color
    linewidth = 0.5)     # can change width of line

plt.xlabel('Release Year')
plt.ylabel('Average Total Streams')
plt.title('Did Listeners Stream Less in 2022?', fontsize = 18, pad=20)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.axvline(
    2022,     # where you want the line to be on the x-axis
    color='#063b0a',
    linewidth = 0.5,
    alpha = 0.7
    )

ax.text(     # # adds text to axis object and specifies where it goes among other attributes
    2022+0.05,
    2225,
    'Dip in Total Streams',
    color='#063b0a',
    fontsize=8
)

plt.ylim(0, 3500)

plt.show()
```

```python
### PIE CHARTS ###

# pie charts can be valuable when you have just a few categories
```

```python
category = 'Platform Type'
values = 'Total Streams (Millions)'

platform_totals = spotify_df.groupby(category)[values].sum()
platform_totals
```

```python
plt.figure(figsize=(8,8))
plt.pie(
    platform_totals,
    labels=platform_totals.index,
    startangle=110,
    colors=['gray', '#1aaa10']
    )

plt.title('Near Even Split Between Premium and Free Users')

# can also add a legend
plt.legend(
    labels = platform_totals.index,
# the loc='best' will automatically put the legend in the 'best' place     
    loc='best',
    title = 'Platform'
)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(
    platform_totals,
    labels=None,
    startangle=110,
    colors=['gray', '#1aaa10'],
    # this will allow you to see the values and it can be formatted to show the values of wedges; here it will show the number and one decimal place for the % 
    autopct='%1.1f%%'
    )

plt.title('Near Even Split Between Premium and Free Users')

plt.legend(
    labels = platform_totals.index,
    loc='lower right',     # can be changed from best as well
    title = 'Platform'
)

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))


wedges, texts, autotexts = ax.pie(
    platform_totals,
    autopct='%1.1f%%',
    startangle=110,
    colors=['gray', '#1aaa10'],
    )

ax.set_title(
    'Near Even Split Between Premium and Free Users',
    fontsize=40,
    pad=15,
    loc='center'
    )

# autotext can be used to change all of the text in the chart
for autotext in autotexts:
    autotext.set_fontsize(30)
    autotext.set_color('white')
    autotext.set_weight('bold')

ax.legend(
    wedges,
    platform_totals.index,
    loc='lower center',
    bbox_to_anchor = (0.5, 0.9),     # bbox allows you to move position very specific to the chart; here it is putting the legend at 0.5 on the x-axis and 0.9 on the y-axis
    ncol=2     # changes the legend to horizontal in this instance
)

plt.tight_layout()
plt.show()
```

```python
### HISTOGRAMS ###

# histograms are good for showing a distribution
```

```python
plt.hist(
    spotify_df['Avg Stream Duration (Min)'],
    bins=25
    # In a pandas chart, bins are intervals used to group continuous numerical data into discrete categories. This process, called "binning," is most commonly seen in histograms and hexagonal bin plots.  #
    # Instead of showing every single data point, binning provides a high-level view of how values are distributed across a dataset.
)

plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(
    spotify_df['Avg Stream Duration (Min)'],
    bins=20,
    color='#2fba28',
    alpha=0.9,
    edgecolor = 'white'
)

plt.xlabel('Average Stream Duration (Minutes)')
plt.ylabel('Number of Tracks')
plt.title('Distribution of Average Stream Duration', fontsize=16, pad=20)

ax = plt.gca()
for spine in ax.spines.values():
    spine.set_linewidth(0.5)
    spine.set_alpha(0.5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```

```python
### BOXPLOTS ###     --- Box Plot / Whisker Plot / Box & Whisker Plot ---

# boxplots are a compact way to compare the spread of a numeric variable across different categories; a way to see the spread and the shape of a dataset
# a boxplot takes in a list of numeric datasets
```

```python
plt.boxplot(
    [spotify_df[spotify_df['Platform Type'] == 'Free']['Skip Rate (%)'],
     spotify_df[spotify_df['Platform Type'] == 'Premium']['Skip Rate (%)']]
)

plt.xticks([1,2],['Free', 'Premium'])     # xticks refers to the tick marks and their corresponding labels displayed on the x-axis of a plot
plt.xlabel('Platform Type')
plt.ylabel('Skip Rate (%)')
plt.title('Skip Rate (%) by Platform Type')

plt.show()
```
